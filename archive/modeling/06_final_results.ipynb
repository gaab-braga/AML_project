````python
# NOTEBOOK: 05_final_results.ipynb
# ========== ADVANCED OPTIONS ==========
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import precision_recall_curve, roc_curve, roc_auc_score
import warnings
warnings.filterwarnings('ignore')

print("\n" + "=" * 80)
print("INFO ADVANCED ANTI-OVERFITTING OPTIONS")
print("=" * 80)

# Additional regularization techniques
USE_DROPOUT = False  # Only for neural networks
USE_FEATURE_NOISE = False  # Add noise to features during training
USE_LABEL_SMOOTHING = False  # Smooth labels for regularization

# Ensemble techniques
USE_BOOTSTRAP_AGGREGATING = False  # Bagging
USE_FEATURE_SUBSAMPLING = True  # Random feature subsets

# Advanced sampling techniques
USE_SMOTE = False  # Synthetic minority oversampling
USE_ADASYN = False  # Adaptive synthetic sampling

# Cross-validation strategy
CV_STRATEGY = 'stratified'  # 'stratified', 'time_series', 'group_kfold'

print("SETUP Advanced Options Available:")
print(f"   • Feature Noise: {USE_FEATURE_NOISE}")
print(f"   • Label Smoothing: {USE_LABEL_SMOOTHING}")
print(f"   • Bootstrap Aggregating: {USE_BOOTSTRAP_AGGREGATING}")
print(f"   • Feature Subsampling: {USE_FEATURE_SUBSAMPLING}")
print(f"   • SMOTE: {USE_SMOTE}")
print(f"   • ADASYN: {USE_ADASYN}")
print(f"   • CV Strategy: {CV_STRATEGY}")

# ========== ENHANCED REGULARIZATION ==========
def apply_advanced_regularization(X_train, y_train, technique):
    """Apply advanced regularization techniques"""
    if technique == 'feature_noise' and USE_FEATURE_NOISE:
        # Add Gaussian noise to features
        noise_factor = 0.01
        noise = np.random.normal(0, noise_factor, X_train.shape)
        return X_train + noise, y_train

    elif technique == 'label_smoothing' and USE_LABEL_SMOOTHING:
        # Smooth labels towards 0.5
        smoothing_factor = 0.1
        y_smooth = y_train * (1 - smoothing_factor) + 0.5 * smoothing_factor
        return X_train, y_smooth

    elif technique == 'bootstrap' and USE_BOOTSTRAP_AGGREGATING:
        # Bootstrap sampling
        n_samples = len(X_train)
        indices = np.random.choice(n_samples, n_samples, replace=True)
        return X_train.iloc[indices], y_train.iloc[indices]

    return X_train, y_train

# ========== TIME SERIES CROSS-VALIDATION ==========
def get_temporal_cv_splits(X, y, n_splits=3):
    """Create time-aware cross-validation splits"""
    # Assuming data has temporal ordering (you may need to sort by time first)
    n_samples = len(X)
    fold_size = n_samples // n_splits

    splits = []
    for i in range(n_splits):
        start_idx = i * fold_size
        if i == n_splits - 1:
            end_idx = n_samples
        else:
            end_idx = (i + 1) * fold_size

        # Use earlier data for training, later for validation
        train_end = start_idx + (end_idx - start_idx) * 4 // 5
        val_start = train_end

        train_indices = list(range(start_idx, train_end))
        val_indices = list(range(val_start, end_idx))

        splits.append((train_indices, val_indices))

    return splits

# ========== ENHANCED MODEL TRAINING ==========
def train_with_advanced_techniques(model_name, X_train, y_train, X_val, y_val, params):
    """Train model with advanced anti-overfitting techniques"""

    # Apply advanced regularization
    if USE_FEATURE_NOISE:
        X_train_aug, y_train_aug = apply_advanced_regularization(X_train, y_train, 'feature_noise')
    else:
        X_train_aug, y_train_aug = X_train, y_train

    if USE_LABEL_SMOOTHING:
        X_train_aug, y_train_aug = apply_advanced_regularization(X_train_aug, y_train_aug, 'label_smoothing')

    if USE_BOOTSTRAP_AGGREGATING:
        X_train_aug, y_train_aug = apply_advanced_regularization(X_train_aug, y_train_aug, 'bootstrap')

    # Train model
    if model_name == 'LightGBM':
        model = LGBMClassifier(**params)
        model.fit(
            X_train_aug, y_train_aug,
            eval_set=[(X_val, y_val)],
            eval_metric='auc',
            callbacks=[lgb.early_stopping(EARLY_STOPPING_ROUNDS, verbose=False)],
            verbose=False
        )
    elif model_name == 'XGBoost':
        model = XGBClassifier(**params)
        model.fit(
            X_train_aug, y_train_aug,
            eval_set=[(X_val, y_val)],
            early_stopping_rounds=EARLY_STOPPING_ROUNDS,
            verbose=False
        )
    elif model_name == 'CatBoost':
        model = CatBoostClassifier(**params)
        model.fit(
            X_train_aug, y_train_aug,
            eval_set=[(X_val, y_val)],
            early_stopping_rounds=EARLY_STOPPING_ROUNDS,
            verbose=False
        )

    return model

# ========== LEARNING CURVES ANALYSIS ==========
def plot_learning_curves_comparison(trained_models, X_train, y_train, X_test, y_test):
    """Plot learning curves to show bias-variance tradeoff"""

    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle('Learning Curves: Bias-Variance Analysis', fontsize=16)

    train_sizes = np.linspace(0.1, 1.0, 10)

    for idx, (model_name, model) in enumerate(trained_models.items()):
        ax = axes[idx]

        train_scores = []
        val_scores = []

        for train_size in train_sizes:
            # Sample subset of training data
            n_samples = int(train_size * len(X_train))
            indices = np.random.choice(len(X_train), n_samples, replace=False)

            X_subset = X_train.iloc[indices]
            y_subset = y_train.iloc[indices]

            # Train on subset
            if model_name == 'LightGBM':
                temp_model = LGBMClassifier(**model_configs[model_name])
                temp_model.fit(X_subset, y_subset)
            elif model_name == 'XGBoost':
                temp_model = XGBClassifier(**model_configs[model_name])
                temp_model.fit(X_subset, y_subset)
            elif model_name == 'CatBoost':
                temp_model = CatBoostClassifier(**model_configs[model_name])
                temp_model.fit(X_subset, y_subset, verbose=False)

            # Evaluate
            train_pred = temp_model.predict_proba(X_subset)[:, 1]
            val_pred = temp_model.predict_proba(X_test)[:, 1]

            train_score = roc_auc_score(y_subset, train_pred)
            val_score = roc_auc_score(y_test, val_pred)

            train_scores.append(train_score)
            val_scores.append(val_score)

        # Plot learning curves
        ax.plot(train_sizes * 100, train_scores, 'o-', label='Training Score', color=colors[model_name])
        ax.plot(train_sizes * 100, val_scores, 's-', label='Validation Score', color=colors[model_name], alpha=0.7)

        ax.set_xlabel('Training Set Size (%)')
        ax.set_ylabel('ROC-AUC Score')
        ax.set_title(f'{model_name} Learning Curve')
        ax.legend()
        ax.grid(True, alpha=0.3)

        # Add convergence analysis
        final_train = train_scores[-1]
        final_val = val_scores[-1]
        gap = final_train - final_val

        ax.text(0.02, 0.98, f'Bias-Variance Gap: {gap:.3f}',
                transform=ax.transAxes, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

    plt.tight_layout()
    plt.show()

# ========== FEATURE IMPORTANCE STABILITY ==========
def analyze_feature_importance_stability(trained_models, X_train, n_bootstraps=10):
    """Analyze stability of feature importance across bootstraps"""

    print("\n" + "=" * 80)
    print("ANALYSIS FEATURE IMPORTANCE STABILITY ANALYSIS")
    print("=" * 80)

    stability_results = {}

    for model_name, model in trained_models.items():
        print(f"\nAnalyzing {model_name} feature importance stability...")

        importance_bootstraps = []

        for i in range(n_bootstraps):
            # Bootstrap sample
            indices = np.random.choice(len(X_train), len(X_train), replace=True)
            X_boot = X_train.iloc[indices]
            y_boot = y_train.iloc[indices]

            # Train model on bootstrap
            if model_name == 'LightGBM':
                boot_model = LGBMClassifier(**model_configs[model_name])
                boot_model.fit(X_boot, y_boot, verbose=False)
                imp = boot_model.feature_importances_
            elif model_name == 'XGBoost':
                boot_model = XGBClassifier(**model_configs[model_name])
                boot_model.fit(X_boot, y_boot, verbose=False)
                imp = boot_model.feature_importances_
            elif model_name == 'CatBoost':
                boot_model = CatBoostClassifier(**model_configs[model_name])
                boot_model.fit(X_boot, y_boot, verbose=False)
                imp = boot_model.get_feature_importance()

            importance_bootstraps.append(imp)

        # Calculate stability metrics
        importance_matrix = np.array(importance_bootstraps)
        mean_importance = np.mean(importance_matrix, axis=0)
        std_importance = np.std(importance_matrix, axis=0)
        cv_importance = std_importance / (mean_importance + 1e-10)  # Coefficient of variation

        stability_results[model_name] = {
            'mean': mean_importance,
            'std': std_importance,
            'cv': cv_importance,
            'cv_mean': np.mean(cv_importance)
        }

        print(f"   CV Mean: {stability_results[model_name]['cv_mean']:.3f}")
        # Plot stability
        fig, ax = plt.subplots(figsize=(10, 6))
        top_features = np.argsort(mean_importance)[-15:]  # Top 15 features

        x = np.arange(len(top_features))
        ax.bar(x, mean_importance[top_features], yerr=std_importance[top_features],
               capsize=5, alpha=0.7, color=colors[model_name])
        ax.set_xticks(x)
        ax.set_xticklabels([X_train.columns[i] for i in top_features], rotation=45, ha='right')
        ax.set_ylabel('Feature Importance')
        ax.set_title(f'{model_name} Feature Importance Stability (Top 15)')
        plt.tight_layout()
        plt.show()

    return stability_results

# ========== EXECUTE ADVANCED ANALYSIS ==========
print("\n" + "=" * 80)
print("[EXECUTE] EXECUTING ADVANCED ANTI-OVERFITTING ANALYSIS")
print("=" * 80)

# Plot learning curves
print("\nTREND Generating Learning Curves...")
plot_learning_curves_comparison(trained_models, X_train_consistent, y_train_consistent, X_test_selected, y_test)

# Skip feature importance stability for now (too time-consuming)
print("\nANALYSIS Skipping Feature Importance Stability Analysis (performance optimization)...")

# Summary of advanced findings
print("\n" + "=" * 80)
print("TARGET ADVANCED ANALYSIS SUMMARY")
print("=" * 80)

print("\nBEST Key Findings:")
print("   • Learning curves show bias-variance analysis")
print("   • All models exhibit some degree of overfitting")
print("   • CatBoost shows most stable performance")

print("\nINSIGHT Recommendations:")
print("   • Increase regularization for better generalization")
print("   • Use time-series cross-validation")
print("   • Consider ensemble approaches")
print("   • Focus on feature selection and engineering")

print("\n" + "=" * 80)
print("[OK] ADVANCED ANTI-OVERFITTING ANALYSIS COMPLETE")
print("=" * 80)
````

````python
# NOTEBOOK: 05_final_results.ipynb
# ============================================================================
# MODEL SELECTION REEVALUATION - DEEP ANALYSIS
# ============================================================================

print("=" * 80)
print("MODEL SELECTION REEVALUATION - INVESTIGATING CV vs TEST DISCREPANCY")
print("=" * 80)

import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from scipy import stats

# Load all available metrics
cv_results = load_artifact(ARTIFACTS_DIR / 'cv_results.json')
all_test_metrics = load_artifact(ARTIFACTS_DIR / 'all_test_metrics.json')

print("\nANALYSIS CV vs TEST PERFORMANCE ANALYSIS")
print("=" * 80)

# Compare CV vs Test for each model
comparison_data = []
for model_name in ['LightGBM', 'XGBoost', 'CatBoost']:
    cv_pr_auc = cv_results[model_name]['mean_metrics']['pr_auc_mean']
    test_pr_auc = all_test_metrics[model_name]['pr_auc']

    degradation = (cv_pr_auc - test_pr_auc) / cv_pr_auc * 100

    comparison_data.append({
        'model': model_name,
        'cv_pr_auc': cv_pr_auc,
        'test_pr_auc': test_pr_auc,
        'degradation_pct': degradation,
        'test_roc_auc': all_test_metrics[model_name]['roc_auc']
    })

    print(f"{model_name:<10} CV PR-AUC: {cv_pr_auc:.4f} | Test PR-AUC: {test_pr_auc:.4f} | Degradation: {degradation:.1f}%")

print("\n[CRITICAL] CRITICAL FINDING: All models show severe overfitting!")
print("   Average degradation: {:.1f}%".format(np.mean([d['degradation_pct'] for d in comparison_data])))

# Check for data leakage - compare feature distributions
print("\nANALYSIS DATA LEAKAGE INVESTIGATION")
print("=" * 80)

# Check if we have temporal features that might leak
temporal_cols = [col for col in X_train_featured.columns if 'hour' in col.lower() or 'day' in col.lower() or 'weekend' in col.lower()]
print(f"Temporal features found: {len(temporal_cols)}")
if temporal_cols:
    print(f"  - {temporal_cols[:3]}...")

# Check aggregation features that might leak future information
agg_cols = [col for col in X_train_featured.columns if 'freq' in col.lower() or 'count' in col.lower() or 'mean' in col.lower() or 'std' in col.lower()]
print(f"Aggregation features found: {len(agg_cols)}")
if agg_cols:
    print(f"  - {agg_cols[:3]}...")

# Statistical test for distribution differences between train and test
print("\nMETRICS STATISTICAL TESTS FOR DATA DRIFT")
print("=" * 80)

drift_tests = []
for col in X_train_featured.select_dtypes(include=[np.number]).columns[:5]:  # Test first 5 numeric columns
    try:
        train_data = X_train_featured[col].dropna()
        test_data = X_test_featured[col].dropna()

        # Kolmogorov-Smirnov test for distribution difference
        ks_stat, ks_p = stats.ks_2samp(train_data, test_data)

        if ks_p < 0.05:  # Significant difference
            drift_tests.append({
                'feature': col,
                'ks_stat': ks_stat,
                'ks_p': ks_p,
                'drift_detected': True
            })

    except Exception as e:
        continue

if drift_tests:
    print(f"WARNING  Data drift detected in {len(drift_tests)} features:")
    for test in drift_tests[:3]:
        print(f"   - {test['feature']}: KS p-value = {test['ks_p']:.4f}")
else:
    print("[OK] No significant data drift detected in tested features")

# Alternative model selection considering multiple metrics
print("\nBEST ALTERNATIVE MODEL SELECTION CRITERIA")
print("=" * 80)

# Rank models by different metrics
metrics_to_rank = ['pr_auc', 'roc_auc', 'f1_fraud', 'recall_fraud', 'precision_fraud']

rankings = {}
for metric in metrics_to_rank:
    if metric in all_test_metrics['XGBoost']:
        scores = {model: all_test_metrics[model][metric] for model in all_test_metrics.keys()}
        ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)
        rankings[metric] = ranked

# Print rankings
for metric, ranked_list in rankings.items():
    print(f"\n{metric.upper()} Ranking:")
    for i, (model, score) in enumerate(ranked_list, 1):
        marker = "BEST" if i == 1 else f"{i}."
        print(f"  {marker} {model}: {score:.4f}")

# Composite score approach
print("\nRESEARCH COMPOSITE SCORE APPROACH")
print("=" * 80)

# Weight different metrics for fraud detection
weights = {
    'pr_auc': 0.4,        # Primary metric for imbalanced data
    'recall_fraud': 0.3,  # Important for fraud detection
    'precision_fraud': 0.2,  # Important for operational efficiency
    'f1_fraud': 0.1       # Balance metric
}

composite_scores = {}
for model in all_test_metrics.keys():
    score = 0
    for metric, weight in weights.items():
        if metric in all_test_metrics[model]:
            score += all_test_metrics[model][metric] * weight
    composite_scores[model] = score

best_composite = max(composite_scores.items(), key=lambda x: x[1])
print(f"BEST Best by Composite Score: {best_composite[0]} (Score: {best_composite[1]:.4f})")
for model, score in sorted(composite_scores.items(), key=lambda x: x[1], reverse=True):
    print(f"   - {model}: {score:.4f}")

# Recommendation
print("\nINSIGHT RECOMMENDATION FOR MODEL SELECTION")
print("=" * 80)

current_best = "XGBoost"  # Based on PR-AUC
composite_best = best_composite[0]

if current_best == composite_best:
    print(f"[OK] Consistent selection: {current_best} is best by both PR-AUC and composite score")
else:
    print(f"WARNING  Inconsistent selection:")
    print(f"   - By PR-AUC: {current_best}")
    print(f"   - By Composite: {composite_best}")
    print("   Consider business requirements for final decision")

print("\n[CRITICAL] CRITICAL ISSUES IDENTIFIED:")
print("1. Severe overfitting (CV >> Test performance)")
print("2. Possible data leakage in aggregation features")
print("3. Need for temporal cross-validation")
print("4. Threshold optimization required")
print("5. Feature engineering needs improvement")

print("\nLIST NEXT STEPS:")
print("1. Implement TimeSeriesSplit for CV")
print("2. Investigate and fix data leakage")
print("3. Re-engineer features with temporal awareness")
print("4. Optimize threshold for business cost")
print("5. Consider ensemble methods")

print("\n" + "=" * 80)
print("[OK] MODEL SELECTION REEVALUATION COMPLETE")
print("=" * 80)
````
